# ep11 merge

```python
import pandas as pd
ver=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
ver1=pd.DataFrame({"a":[1,2,3,4],"b":[21,22,23,24]})

```

```python
pd.merge(ver,ver1,on="a")
```

```python
pd.merge(ver1,ver,on="a")
```

```python
ver1=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
ver2=pd.DataFrame({"a":[1,2,3,5],"c":[21,22,23,24]})
pd.merge(ver1,ver2,on="a")
```

```python
pd.merge(ver2,ver1,on="a")
```

```python
pd.merge(ver1,ver2,how="inner")
```

```python
pd.merge(ver1,ver2,how="left")
```

```python
pd.merge(ver1,ver2,how="right")
```

```python
pd.merge(ver1,ver2,how="outer",indicator=True)
```

# join

```python
import pandas as pd
ver=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]})
ver1=pd.DataFrame({"c":[10,11,12,13],"d":[15,16,17,18]})
ver1.join(ver)
```

```python
ver.join(ver1)
```

```python
ver=pd.DataFrame({"a":[1,2,3,4],"b":[11,12,13,14]},index=["a","b","c","d"])
ver2=pd.DataFrame({"c":[10,11],"d":[15,16]},index=["a","b"])
print(ver)
print()
print(ver2)
```

```python
ver1.join(ver,how="left")
```

```python
ver1.join(ver,how="right")
```

```python
ver.join(ver1,how="inner")
```

```python
ver.join(ver1,how="outer")
```
